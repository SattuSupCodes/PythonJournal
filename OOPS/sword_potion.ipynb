```python
class Item:
  def __init__(self, name, value):
    self.value = value
    self.name = name
  def use(self):
    print("You cannot use this item.")
class Potion(Item):
  def use(self):
    print("you are using potion")
class Weapon(Item):
  def use(self):
    print("you are using weapon")
class Player:
  def __init__(self,name, health):
    self.name = name
    self.health = health
    self.inventory = []
  def add_item(self,item):
    self.inventory.append(item)
  def use_item(self,item_name):
    for item in self.inventory:
      if item.name == item_name:
        item.use()
        return
      print("no item found")
  def show_inventory(self):
    print("Inventory:")
    for item in self.inventory:
      print("-", item.name)
```

```python
player = Player("Sattu", 100)
```

```python
potion = Potion("health potion", 50)
sword = Weapon("sword", 30)
```

```python
player.add_item(potion)
player.add_item(sword)
```

```python
player.show_inventory()
```
