```python
class Notifications:
  def __init__(self, sender, receiver, message):
    self.sender = sender
    self.receiver = receiver
    self.message = message
  def send_message(self,sender, receiver , message):
    print(f"from {sender} to {receiver}: {message}")
class Email_Notif(Notifications):
  def send_message(self, sender, receiver, message):
    print(f"from {sender} to {receiver}: Email: {message}")
class SMS_notif(Notifications):
  def send_message(self, sender, receiver, message):
    print(f"from {sender} to {receiver}: SMS: {message}")
```
