#Bank Class

```python
class Bank:
  def __init__(self, name, balance):
    self.name = name
    self.balance = balance
  def deposite(self, amount):
    self.balance += amount
    return self.balance
  def withdraw(self, amount):
    if amount<=self.balance:
      self.balance -= amount
      return self.balance
    else:
      return False
```

```python
bank = Bank("Sattu", 10000)
bank.deposite(500)
bank.withdraw(200000)

#Wow yaar stsatuu
```

#Puzzle: Shape Mystery

Goal: To create a shape class to find area of shapes

Twist: The class should itself know when to use circle formula or rectangle forumla without hard coding if statments


```python
class Shape:
  def __init__(self, shape_name):
    self.name = shape_name
  def area(self):
    pass


```

```python
class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius
  def area(self):
    return 3.14*self.radius**2
```

```python
class Rectangle(Shape):
  def __init__(self, width,length):
    self.length = length
    self.width = width
  def area(self):
    return self.length*self.width
```

```python
#poly smt shit
shapes =[
    Circle(7),
    Rectangle(6,8)
]
for shape in shapes:
  print(shape.area())
```
