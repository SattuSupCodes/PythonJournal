#RANDOM TENSORS
They are important because the way many neural networks learn is that they start with tensors full of randomnumbers and then adjust those random numbers to better represent the data.

```start with random numbers -> look at data -> update random numbers -> look at data -> update ```

````python
random_tensor = torch.rand(3,4) #SIZE (3,4)

print(random_tensor)
````

````python
random_image_size_tensor = torch.rand(size=(3,224,224) )#height, width, color channel
print(random_image_size_tensor)
````

````python
#creating a range of tensors and tensors like

torch.arange(0,10)
````
