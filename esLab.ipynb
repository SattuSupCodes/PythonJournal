```python
facts = {"eats_protein", "builds_muscle"}

rules = [
    ({"eats_protein", "builds_muscle"}, "healthy_person"),
    ({"healthy_person"}, "eligible_for_army"),
    ({"eligible_for_army"}, "can_serve_country")
]

def forward_chaining(facts,rules):
  flag = True
  while flag:
    flag = False
    for condition, conclusion in rules:
      if condition.issubset(facts) and conclusion not in facts:
        facts.add(conclusion)
        print("new fact inferred:", conclusion)
        flag = True
  return facts
final_facts = forward_chaining(facts,rules)
print("\nFinal Facts:")
for f in final_facts:
  print("-", f)

```

```python
facts = {
    "attends_classes",
    "submits_assignments",
    "studies_daily",
    "has_good_attendance",
    "clears_internal_assessment"
}

rules = [

    ({"attends_classes", "submits_assignments"}, "regular_student"),
    ({"regular_student", "studies_daily"}, "well_prepared"),
    ({"has_good_attendance", "clears_internal_assessment"}, "eligible_for_exam"),
    ({"eligible_for_exam"}, "can_sit_exam"),
    ({"well_prepared", "can_sit_exam"}, "can_pass_exam"),


    ({"can_pass_exam"}, "gets_good_grades"),
    ({"gets_good_grades"}, "eligible_for_scholarship"),

    ({"eligible_for_scholarship"}, "financial_support"),
    ({"gets_good_grades", "financial_support"}, "academic_success"),
]
def forward_chaining(facts, rules):
    flag = True
    while flag:
        flag = False
        for condition, conclusion in rules:
            if condition.issubset(facts) and conclusion not in facts:
                facts.add(conclusion)
                print("New fact inferred:", conclusion)
                flag = True
    return facts

final_facts = forward_chaining(facts, rules)

print("\nFinal Facts:")
for f in final_facts:
    print("-", f)

```

```python
import sqlite3

# Connect to database (creates file if not exists)
conn = sqlite3.connect("college.db")
cursor = conn.cursor()

# Create Students table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Students (
    student_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    department TEXT,
    year INTEGER
)
""")

# Create Attendance table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Attendance (
    attendance_id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_id INTEGER,
    attendance_percentage REAL,
    FOREIGN KEY (student_id) REFERENCES Students(student_id)
)
""")

# Create Internal Assessment table
cursor.execute("""
CREATE TABLE IF NOT EXISTS InternalAssessment (
    assessment_id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_id INTEGER,
    internal_marks INTEGER,
    FOREIGN KEY (student_id) REFERENCES Students(student_id)
)
""")

conn.commit()

# Insert a student
cursor.execute("INSERT INTO Students (name, department, year) VALUES (?, ?, ?)",
               ("Rahul Sharma", "CSE", 2))

# Get student id
student_id = cursor.lastrowid

# Insert attendance
cursor.execute("INSERT INTO Attendance (student_id, attendance_percentage) VALUES (?, ?)",
               (student_id, 82.5))

# Insert internal marks
cursor.execute("INSERT INTO InternalAssessment (student_id, internal_marks) VALUES (?, ?)",
               (student_id, 40))

conn.commit()


cursor.execute("""
SELECT s.name,
       CASE
           WHEN a.attendance_percentage >= 75 AND i.internal_marks >= 25
           THEN 'Eligible'
           ELSE 'Not Eligible'
       END AS eligibility
FROM Students s
JOIN Attendance a ON s.student_id = a.student_id
JOIN InternalAssessment i ON s.student_id = i.student_id
""")

results = cursor.fetchall()

for row in results:
    print("Name:", row[0])
    print("Exam Status:", row[1])
conn.close()
```
