# Decorators In Python

```python
# A function that extends the behavior of another function without modifying the base function
def add_sprinkles(func):
  def wrapper(*args, **kwargs): #you have to have this
    print("you add sprinkles")
    func(*args, **kwargs)
  return wrapper
def addfudge(func):
  def wrapper(*args, **kwargs):
    print("added fudge")
    func(*args, **kwargs)
  return wrapper

@add_sprinkles
@addfudge
def get_icecream(flavor):
  print(f"your icecream {flavor}")
get_icecream("chocolate")



```
